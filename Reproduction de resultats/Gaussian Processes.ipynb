### Bayesian Optimization Tutorial

On va estimer un processus Gaussien à l'aide de méthodes bayésiennes : Vanilla test

```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#### Génération de notre échantillon 
sigma_eps = 0.1
X = np.linspace(-1,1,100)[:,np.newaxis]
Epsilon = np.random.normal(scale=sigma_eps, size=100)[:,np.newaxis]

### Définition du modèle
Y = np.power(X,3) + Epsilon
```

```python
fig, ax = plt.subplots()
ax.plot(X[:,0],Y[:,0],'go')
ax.set_xlabel("X",fontsize=13)
ax.set_ylabel("Y",fontsize=13)
ax.set_title("Echantillon",fontsize=15)
plt.show()
```

```python
### Génération du premier processus gaussien
#1) Définition du Kernel i.e la fonction pour estimer 
#la matrice de var-covar du processus

def ker(a,b,param):
    sqr_dist= np.square(a.reshape(1,-1)-b.reshape(1,-1))
    return( np.exp(-0.5 * sqr_dist / param))
```

```python
param=0.2
K = ker(X,X,param=param)
```

```python
import numpy as np
import matplotlib.pyplot as pl

# Test data
n = 50
Xtest = np.linspace(-5, 5, n).reshape(-1,1)

# Define the kernel function
def kernel(a, b, param):
    # Let's remind that if x.shape= (1,K) and y.shape = (K,), (x+y).shape = (K,K) and (x+y)_ij=(x_i + y_j) 
    # See np.array([1,2,3]).reshape(3,1) + np.array([1,2,3])
    sqdist = (a**2).reshape(-1,1) + np.sum(b**2,1) - 2*np.dot(a, b.T)
    return np.exp(-.5 * (1/param) * sqdist)

param = 0.1
K_ss = kernel(Xtest, Xtest, param)

# Get cholesky decomposition (square root) of the
# covariance matrix
L = np.linalg.cholesky(K_ss + 1e-15*np.eye(n))
# Sample 3 sets of standard normals for our test points,
# multiply them by the square root of the covariance matrix
f_prior = np.dot(L, np.random.normal(size=(n,3)))

# Now let's plot the 3 sampled functions.
pl.plot(Xtest, f_prior)
pl.axis([-5, 5, -3, 3])
pl.title('Three samples from the GP prior')
pl.show()
```

```python
# Noiseless training data
Xtrain = np.array([-4, -3, -2, -1, 1]).reshape(5,1)
ytrain = np.sin(Xtrain)

# Apply the kernel function to our training points
K = kernel(Xtrain, Xtrain, param)
L = np.linalg.cholesky(K )

# Compute the mean at our test points.
K_s = kernel(Xtrain, Xtest, param)
Lk = np.linalg.solve(L, K_s) #Solve Lx = K_s
mu = np.dot(Lk.T, np.linalg.solve(L, ytrain)).reshape((n,))

# Compute the standard deviation so we can plot it
s2 = np.diag(K_ss) - np.sum(Lk**2, axis=0)
stdv = np.sqrt(s2)
# Draw samples from the posterior at our test points.
L = np.linalg.cholesky(K_ss  - np.dot(Lk.T, Lk))

#on a la moyenne mu et la décomposition de Cholesky de la variance L.
#On peut donc demander 3 échantillons E1, E2, E3 ~ N(0,In) à numpy 
# et mu + L * E1 ~ GP(mu , L L^T) ! i.e un échantillon de la loi posterior
f_post = mu.reshape(-1,1) + np.dot(L, np.random.normal(size=(n,3))) 

pl.plot(Xtrain, ytrain, 'bs', ms=8) #Echantillon
pl.plot(Xtest, f_post, label="posterior")  
pl.plot(Xtest,np.sin(Xtest),label='sin')
pl.gca().fill_between(Xtest.flat, mu-2*stdv, mu+2*stdv, color="#dddddd")
pl.plot(Xtest, mu, 'r--', lw=2)
pl.axis([-5, 5, -3, 3])
pl.legend()
pl.title('Three samples from the GP posterior')
pl.show()
```
